```python
import csv
#éœ€è¦å£°æ˜å…¨å±€å˜é‡
#æ”¯å‡ºè®°å½•
paymentlist=[]
#æ”¶å…¥è®°å½•
incomelist=[]
#æ”¯å‡ºå¤šå°‘ç¬”æ•°æ®
paymentnum=0
#æ”¶å…¥å¤šå°‘ç¬”æ•°æ®
incomenum=0
def read_csv(filename):
    csv_file = csv.reader(open(filename))
    count = 0
    for line in csv_file:
        if count >= 17:
            # print(line[0])
            # æ‰€æœ‰çš„è¯¦ç»†ä¿¡æ¯
            #åµŒå¥—å‡½æ•°
            savedata(line)

        count += 1
def savedata(line):
    global paymentnum
    global incomenum
    #å¤„ç†åŒºåˆ†æ”¶å…¥å’Œæ”¯å‡ºï¼ŒæŠŠæ•°æ®æºåŒºåˆ†å¼€
    if line[4]=='æ”¯å‡º':
        paymentlist.append(line)
        paymentnum+=1
    else:
        incomelist.append(line)
        incomenum+=1
    #åˆ—è¡¨æ˜¯èƒ½ç›´æ¥è¿½åŠ æ•°æ®çš„  ä½†æ˜¯æ•°å­—ä¸èƒ½ç›´æ¥è¿½åŠ æ•°æ®
```

```python


#æ”¯å‡ºæ±‚å’Œ
def getPayMentList():
    #æ€è·¯
    #1ã€éå†æ”¶å…¥åˆ—è¡¨
    #2ã€æ¯è¡Œå†…å®¹é‡Œé¢ï¼Œè·å¾—å¤šå°‘é’±
    #3ã€æ±‚å’Œçš„æ“ä½œ
    all=0
    for item in paymentlist:
        all+=float(item[5][1:])
    print(all)
    return all
#æ”¶å…¥æ±‚å’Œ
def getInConmeList():
    all = 0
    for item in incomelist:
        all += float(item[5][1:])
    print(all)
    return all

#æ”¯å‡ºæ˜ç»†å½’çº³æ±‡æ€»
def getPayMentListDetail():
    #ä¿å­˜åœ¨å­—å…¸ä¸­
    dict1={}
    for item in paymentlist:
        x=dict1.get(item[2])
        if x==None:
            #æ²¡æœ‰è¿™ä¸ªkeyï¼Œå¯¹å­—å…¸è¿›è¡Œèµ‹å€¼
            dict1[item[2]]=float(item[5][1:])
        else:
            #å­—å…¸ä¸­æœ‰keyå¯¹åº”çš„value  å–å‡ºvalue+ç°åœ¨çš„å€¼
            dict1[item[2]]=dict1[item[2]]+float(item[5][1:])

    #å¦‚æœåœ¨è¿™ä¸ªåŸºç¡€ä¸Šï¼Œæ±‚å¾—æœ€å¤šç»™äº†è°
    #dict={'åº†ä¸°åŒ…å­é“º':10000}
    max=0
    maxtitle=""
    for i in dict1:
        #éå†å­—å…¸å®é™…ä¸Šæ˜¯å¾—åˆ°å­—å…¸çš„key
        temp=dict1[i]
        if max <temp:
            max=temp
            maxtitle=i
    print(max,maxtitle)
    return dict1
```

```python
payMentall=getPayMentList()
incomeall=getInConmeList()
dict1=getPayMentListDetail()
```
