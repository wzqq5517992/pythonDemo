```python
import csv
#需要声明全局变量
#支出记录
paymentlist=[]
#收入记录
incomelist=[]
#支出多少笔数据
paymentnum=0
#收入多少笔数据
incomenum=0
def read_csv(filename):
    csv_file = csv.reader(open(filename))
    count = 0
    for line in csv_file:
        if count >= 17:
            # print(line[0])
            # 所有的详细信息
            #嵌套函数
            savedata(line)

        count += 1
def savedata(line):
    global paymentnum
    global incomenum
    #处理区分收入和支出，把数据源区分开
    if line[4]=='支出':
        paymentlist.append(line)
        paymentnum+=1
    else:
        incomelist.append(line)
        incomenum+=1
    #列表是能直接追加数据的  但是数字不能直接追加数据
```

```python
read_csv("微信账单数据源.csv")
```

```python


#支出求和
def getPayMentList():
    #思路
    #1、遍历收入列表
    #2、每行内容里面，获得多少钱
    #3、求和的操作
    all=0
    for item in paymentlist:
        all+=float(item[5][1:])
    print(all)
    return all
#收入求和
def getInConmeList():
    all = 0
    for item in incomelist:
        all += float(item[5][1:])
    print(all)
    return all

#支出明细归纳汇总
def getPayMentListDetail():
    #保存在字典中
    dict1={}
    for item in paymentlist:
        x=dict1.get(item[2])
        if x==None:
            #没有这个key，对字典进行赋值
            dict1[item[2]]=float(item[5][1:])
        else:
            #字典中有key对应的value  取出value+现在的值
            dict1[item[2]]=dict1[item[2]]+float(item[5][1:])

    #如果在这个基础上，求得最多给了谁
    #dict={'庆丰包子铺':10000}
    max=0
    maxtitle=""
    for i in dict1:
        #遍历字典实际上是得到字典的key
        temp=dict1[i]
        if max <temp:
            max=temp
            maxtitle=i
    print(max,maxtitle)
    return dict1
```

```python
payMentall=getPayMentList()
incomeall=getInConmeList()
dict1=getPayMentListDetail()
```

```python
from pyecharts import Pie

attr = ["收入","支出"]
v1 = [incomeall, payMentall]
pie = Pie("饼图示例")
pie.add("", attr, v1, is_label_show=True)
pie
```

```python
from pyecharts import Pie
#list是转换为列表
attr = list(dict1.keys())
v1 = list(dict1.values())
pie = Pie("饼图示例")
pie.add("", attr, v1, is_label_show=True)
pie
```
